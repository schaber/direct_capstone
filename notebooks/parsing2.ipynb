```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import glob
import codecs
```

```python
name_list = glob.glob("Britol_20/*.html")
```

```python
html_stuff =[]
for i in name_list:
    op=codecs.open(i,'r')
    soup = BeautifulSoup (op, 'lxml')
    html_stuff.append(soup)
```

```python
len(html_stuff)
```

```python
def parsing_files(tlist):
    chem = tlist.find('table')
    chem_name = chem.tbody.tr.b.text
#     print(chem_name)
    chem_info = chem.tbody.tr.td
    list_temp=[]

    for div in chem_info.find_all('div'): 
        temp =  div.dl.dd
        for table in temp.find_all('table'): 
            temp_1 = table.tbody.tr
        #         list_temp = []
            for td in temp_1.find_all('td'): 
                list_temp.append(td.text)
    temp_properties2 = {}

    for t_in in range(0, len(list_temp), 2):
        temp_properties2[list_temp[t_in]] = list_temp[t_in + 1]
    return temp_properties2, chem_name
```

```python
prop_list=[]
for i in html_stuff:
    rty=parsing_files(i)
    prop_list.append(rty)
```

```python
chem_list = []
for i in range(0,100):
    name = prop_list[i][1]
    chem_list.append(name)
```

```python
output1 = pd.DataFrame()
for i in range(0,100):
    output1 = output1.append(prop_list[i][0], ignore_index=True)
```
